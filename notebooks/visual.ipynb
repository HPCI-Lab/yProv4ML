# ProvJSON Visualization

##### Importing libraries and defining constants

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
file_path = "../prov/MetricsType.CSV_False_0/prov_MetricsType.CSV_False_GR0_0.json"
json_data = json.load(open(file_path, "r"))

METRIC = "yProv4ML:cpu_usage"
METRIC2 = "yProv4ML:memory_usage"
```

#### Get metrics as dataframe and plot

```python
df1 = pd.read_csv("../" + json_data["entity"][METRIC]["yProv4ML:path"])
df2 = pd.read_csv("../" + json_data["entity"][METRIC2]["yProv4ML:path"])
df1.columns = ["epoch", "value", "time"]
df2.columns = ["epoch", "value", "time"]

print(df2)
```

```python
sns.set_theme()
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Time (Minutes)')
ax1.set_ylabel("Train Loss (MSE)", color=color)
sns.lineplot(data=df1, x="time", y="value", ax=ax1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel("GPU Power Usage (W)", color=color)
sns.lineplot(data=df2, x="time", y="value", ax=ax2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```
